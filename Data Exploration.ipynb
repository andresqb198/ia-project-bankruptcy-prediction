## Inspeccionando variables categÃ³ricas

```python
categorical_columns = [i for i in df.columns if not i in df._get_numeric_data()]
print(categorical_columns)
```

```python
for c in categorical_columns:
    print ("%10s"%c, np.unique(df[c].dropna()))
```

```python
for column in categorical_columns:
    print(df[column].value_counts())
```

```python
plt.figure(figsize=(20,8))
for i,c in enumerate(categorical_columns):    
    plt.subplot(3,5,i+1)
    k=df[[c,"Bankrupt?"]].dropna()
    for v in df[c].dropna().unique():
        sns.histplot(k['Bankrupt?'][k[c]==v], label=v,kde=True, stat="density", linewidth=0);
        plt.title(c)
    plt.yticks([])
    plt.legend()
    plt.subplot(3,5,i+5)
    vc = k[c].value_counts()
    sns.barplot(vc.index, vc.values)    
    plt.xticks(range(len(vc)), vc.index, rotation="vertical")
    plt.tight_layout()
```
